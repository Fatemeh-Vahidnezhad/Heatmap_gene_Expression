```python
df1 = hk_genes.merge(spl_1[['Gene', 'ZaKh-L_S']], 
                    on = 'Gene', how = 'left').merge(spl_2[['Gene','ZaKh-N_S8']],
                    on = 'Gene', how = 'left').merge(spl_3[['Gene', 'AL-Kh-N_S6']],
                    on = 'Gene', how = 'left').merge(spl_4[['Gene', 'Ar-Kh-W_S18']],
                    on = 'Gene', how = 'left').merge(ctrlz[['Gene', 'ctrlz']],
                     on = 'Gene', how = 'left').merge(ctrlb[['Gene', 'ctrlb']],
                     on = 'Gene', how = 'left')

df1 = df1.dropna(axis = 0)
df1
```

# Normalizing the dataset

```python
def average (lst):
    return sum(lst)/len(lst)



def normalize_data (*args):
    sum_expression = []
    norm_fact = []
    
    #sumation of gene expression for every sample:
    for arg in args:
        a = arg.sum()
        sum_expression.append(a)
    
    #average of gene experssion:
    avg = average(sum_expression)
    
    
    #norm_fact:
    for i in sum_expression:
        norm_fact.append(i/avg)
        
        
    return sum_expression  , norm_fact , avg
        
       

a = df1['ZaKh-L_S']
b = df1['ZaKh-N_S8']
c = df1['AL-Kh-N_S6']
d = df1['Ar-Kh-W_S18']
e = df1['ctrlb']
f = df1['ctrlz']
sum_expression  , norm_fact , avg = normalize_data(a,b,c,d,e,f)
#print(sum_expression  , norm_fact , avg)
```

```python
name_samples = ['ZaKh-L_S','ZaKh-N_S8','AL-Kh-N_S6','Ar-Kh-W_S18','ctrlb','ctrlz' ]


dic = {k:v for k,v in zip(name_samples,norm_fact)}
for key, value in dic.items():
    print(key,value)
    

    
new_df = df1.set_index('Gene')
normalized_data = new_df.div(dic)
normalized_data.head()
```
